### We each need to run below to get the dataset locally

Takes like 15 minutes to run

```python
import kaggle
import os

dataset_name = "ashery/chexpert"
download_path = "./CheXpert-v1.0-small"

# Create destination folder
os.makedirs(download_path, exist_ok=True)

# Download + unzip the dataset
kaggle.api.dataset_download_files(dataset_name, path=download_path, unzip=True)

print(f"Dataset downloaded to: {download_path}")
print("Files in folder:", os.listdir(download_path))
```
